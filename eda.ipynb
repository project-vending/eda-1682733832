```
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```
# loading the dataset from the provided link
df = pd.read_csv("https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv")

```

```
# displaying top 5 rows of the dataset
print("First five rows of the dataset:")
print(df.head())

```

```
# displaying last 5 rows of the dataset
print("\nLast five rows of the dataset:")
print(df.tail())

```

```
# checking for null values
print("\nChecking for null values in the dataset:")
print(df.isnull().sum())

```

```
# displaying basic statistics of the dataset
print("\nBasic statistics of the dataset:")
print(df.describe())

```

```
# ploting the distribution of each predictor varaible using box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop(['Sales'], axis=1))
plt.show()

```

```
# ploting the distribution of Sales variable using histogram
plt.figure(figsize=(10,6))
sns.histplot(df['Sales'], kde=True)
plt.show()

```

```
# ploting pairwise relationships amoung predictors and Sales variable
plt.figure(figsize=(10,6))
sns.pairplot(df)
plt.show()
```
